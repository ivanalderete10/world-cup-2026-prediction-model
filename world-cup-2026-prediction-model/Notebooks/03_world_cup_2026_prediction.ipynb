```python
def win_probability(team_a_rating, team_b_rating):

    return 1 / (
        1 + 10 ** (
            (team_b_rating - team_a_rating) / 400
        )
    )
```

```python
def simulate_match(team_a, team_b):

    rating_a = ranking.loc[
        ranking["team"] == team_a,
        "rating"
    ].iloc[0]

    rating_b = ranking.loc[
        ranking["team"] == team_b,
        "rating"
    ].iloc[0]

    p = win_probability(
        rating_a,
        rating_b
    )

    r = random.random()

    if r < p:
        return team_a
    else:
        return team_b
```

```python
def simulate_tournament(teams):

    quarterfinalists = teams.copy()

    qf_winners = []

    for i in range(0,8,2):

        winner = simulate_match(
            quarterfinalists[i],
            quarterfinalists[i+1]
        )

        qf_winners.append(winner)

    sf_winners = []

    for i in range(0,4,2):

        winner = simulate_match(
            qf_winners[i],
            qf_winners[i+1]
        )

        sf_winners.append(winner)

    champion = simulate_match(
        sf_winners[0],
        sf_winners[1]
    )

    return champion
```

```python
def score_matrix(team_a, team_b, max_goals=5):

    score_a = wc_model.loc[
        wc_model["team"] == team_a,
        "power_score_v3"
    ].iloc[0]

    score_b = wc_model.loc[
        wc_model["team"] == team_b,
        "power_score_v3"
    ].iloc[0]

    xg_a, xg_b = expected_goals(
        score_a,
        score_b
    )

    results = []

    for goals_a in range(max_goals + 1):

        for goals_b in range(max_goals + 1):

            prob = (
                poisson.pmf(goals_a, xg_a)
                *
                poisson.pmf(goals_b, xg_b)
            )

            results.append([
                goals_a,
                goals_b,
                prob
            ])

    df = pd.DataFrame(
        results,
        columns=[
            "team_a_goals",
            "team_b_goals",
            "probability"
        ]
    )

    return df.sort_values(
        "probability",
        ascending=False
    )
```

```python
def match_outcome_probabilities(team_a, team_b):

    matrix = score_matrix(team_a, team_b)

    win_a = matrix[
        matrix["team_a_goals"] >
        matrix["team_b_goals"]
    ]["probability"].sum()

    draw = matrix[
        matrix["team_a_goals"] ==
        matrix["team_b_goals"]
    ]["probability"].sum()

    win_b = matrix[
        matrix["team_a_goals"] <
        matrix["team_b_goals"]
    ]["probability"].sum()

    return win_a, draw, win_b
```

```python
def simulate_match(team_a, team_b):

    matrix = score_matrix(team_a, team_b)

    sampled = matrix.sample(
        n=1,
        weights="probability"
    )

    goals_a = int(sampled["team_a_goals"].iloc[0])
    goals_b = int(sampled["team_b_goals"].iloc[0])

    return goals_a, goals_b
```

```python
def simulate_knockout_match(team_a, team_b):

    goals_a, goals_b = simulate_match(
        team_a,
        team_b
    )

    if goals_a > goals_b:
        winner = team_a

    elif goals_b > goals_a:
        winner = team_b

    else:

        score_a = wc_model.loc[
            wc_model["team"] == team_a,
            "power_score_v3"
        ].iloc[0]

        score_b = wc_model.loc[
            wc_model["team"] == team_b,
            "power_score_v3"
        ].iloc[0]

        prob_a = score_a / (score_a + score_b)

        winner = np.random.choice(
            [team_a, team_b],
            p=[prob_a, 1 - prob_a]
        )

    return winner
```

```python
def simulate_quarterfinals():

    qf_matches = [
        ("Argentina", "Japan"),
        ("Spain", "Senegal"),
        ("France", "Portugal"),
        ("Brazil", "England")
    ]

    semifinalists = []

    for team_a, team_b in qf_matches:

        winner = simulate_knockout_match(
            team_a,
            team_b
        )

        semifinalists.append(winner)

    return semifinalists
```

```python
def simulate_semifinals(semifinalists):

    sf1 = simulate_knockout_match(
        semifinalists[0],
        semifinalists[1]
    )

    sf2 = simulate_knockout_match(
        semifinalists[2],
        semifinalists[3]
    )

    return sf1, sf2
```

```python
def simulate_final(team_a, team_b):

    return simulate_knockout_match(
        team_a,
        team_b
    )
```

```python
def simulate_mini_world_cup():

    semifinalists = simulate_quarterfinals()

    finalist_1, finalist_2 = simulate_semifinals(
        semifinalists
    )

    champion = simulate_final(
        finalist_1,
        finalist_2
    )

    return {
        "semifinalists": semifinalists,
        "finalists": [
            finalist_1,
            finalist_2
        ],
        "champion": champion
    }
```

```python
SCORE_COL = "power_score_v4"
```

```python
def expected_goals(team_a_score, team_b_score):

    base_goals = 1.35
    diff = team_a_score - team_b_score

    xg_a = base_goals + (diff * 1.5)
    xg_b = base_goals - (diff * 1.5)

    xg_a = max(xg_a, 0.2)
    xg_b = max(xg_b, 0.2)

    return xg_a, xg_b


def score_matrix(team_a, team_b, max_goals=5):

    score_a = wc_model.loc[
        wc_model["team"] == team_a,
        SCORE_COL
    ].iloc[0]

    score_b = wc_model.loc[
        wc_model["team"] == team_b,
        SCORE_COL
    ].iloc[0]

    xg_a, xg_b = expected_goals(score_a, score_b)

    results = []

    for goals_a in range(max_goals + 1):
        for goals_b in range(max_goals + 1):

            prob = (
                poisson.pmf(goals_a, xg_a)
                *
                poisson.pmf(goals_b, xg_b)
            )

            results.append([goals_a, goals_b, prob])

    df = pd.DataFrame(
        results,
        columns=[
            "team_a_goals",
            "team_b_goals",
            "probability"
        ]
    )

    return df.sort_values(
        "probability",
        ascending=False
    )


def match_outcome_probabilities(team_a, team_b):

    matrix = score_matrix(team_a, team_b)

    win_a = matrix[
        matrix["team_a_goals"] > matrix["team_b_goals"]
    ]["probability"].sum()

    draw = matrix[
        matrix["team_a_goals"] == matrix["team_b_goals"]
    ]["probability"].sum()

    win_b = matrix[
        matrix["team_a_goals"] < matrix["team_b_goals"]
    ]["probability"].sum()

    return win_a, draw, win_b


def show_match_prediction(team_a, team_b):

    win_a, draw, win_b = match_outcome_probabilities(
        team_a,
        team_b
    )

    print(f"\n{team_a} vs {team_b}\n")

    print(f"{team_a}: {win_a:.1%}")
    print(f"Empate: {draw:.1%}")
    print(f"{team_b}: {win_b:.1%}")

    print("\nTop 5 resultados:\n")

    top5 = score_matrix(team_a, team_b).head(5)

    for _, row in top5.iterrows():
        print(
            f"{int(row['team_a_goals'])}-{int(row['team_b_goals'])}"
            f" → {row['probability']:.1%}"
        )


def simulate_match(team_a, team_b):

    matrix = score_matrix(team_a, team_b)

    sampled = matrix.sample(
        n=1,
        weights="probability"
    )

    goals_a = int(sampled["team_a_goals"].iloc[0])
    goals_b = int(sampled["team_b_goals"].iloc[0])

    return goals_a, goals_b


def simulate_knockout_match(team_a, team_b):

    goals_a, goals_b = simulate_match(team_a, team_b)

    if goals_a > goals_b:
        winner = team_a

    elif goals_b > goals_a:
        winner = team_b

    else:
        score_a = wc_model.loc[
            wc_model["team"] == team_a,
            SCORE_COL
        ].iloc[0]

        score_b = wc_model.loc[
            wc_model["team"] == team_b,
            SCORE_COL
        ].iloc[0]

        prob_a = score_a / (score_a + score_b)

        winner = np.random.choice(
            [team_a, team_b],
            p=[prob_a, 1 - prob_a]
        )

    return winner
```

```python
import pandas as pd
import numpy as np
from pathlib import Path
from math import exp, factorial

print("Regenerando simulador de partidos con 48 selecciones...")

# Buscar archivo de ranking
search_roots = [
    Path.cwd(),
    Path.cwd().parent,
    Path.cwd().parent.parent
]

ranking_files = []

for root in search_roots:
    if root.exists():
        ranking_files.extend(list(root.rglob("world_cup_live_power_ranking.xlsx")))

ranking_files = list(dict.fromkeys(ranking_files))

if not ranking_files:
    raise FileNotFoundError("No encontré world_cup_live_power_ranking.xlsx")

ranking_file = ranking_files[0]

print("Ranking encontrado en:")
print(ranking_file)

ranking = pd.read_excel(ranking_file)

print("Equipos en ranking:", ranking["team"].nunique())

# Validar columnas necesarias
required_cols = [
    "team",
    "team_display",
    "flag",
    "confederation",
    "power_score_live",
    "rating"
]

missing = [c for c in required_cols if c not in ranking.columns]

if missing:
    raise ValueError(f"Faltan columnas en el ranking: {missing}")

# Quedarnos solo con 48 equipos
teams = ranking[
    [
        "team",
        "team_display",
        "flag",
        "confederation",
        "power_score_live",
        "rating"
    ]
].copy()

teams = teams.drop_duplicates(subset=["team"]).reset_index(drop=True)

# Función Poisson
def poisson_prob(k, lam):
    return (lam ** k) * exp(-lam) / factorial(k)

# Simular un partido
def simulate_match(team_a_row, team_b_row, max_goals=5):
    team_a = team_a_row["team"]
    team_b = team_b_row["team"]

    team_a_display = team_a_row["team_display"]
    team_b_display = team_b_row["team_display"]

    power_a = float(team_a_row["power_score_live"])
    power_b = float(team_b_row["power_score_live"])

    rating_a = float(team_a_row["rating"])
    rating_b = float(team_b_row["rating"])

    # Ventaja relativa combinando Power Score y Elo
    advantage = (
        1.20 * (power_a - power_b)
        + (rating_a - rating_b) / 600
    )

    # Goles esperados aproximados
    base_goals = 1.25

    expected_goals_a = base_goals * np.exp(advantage)
    expected_goals_b = base_goals * np.exp(-advantage)

    expected_goals_a = float(np.clip(expected_goals_a, 0.25, 3.50))
    expected_goals_b = float(np.clip(expected_goals_b, 0.25, 3.50))

    rows = []

    for goals_a in range(max_goals + 1):
        for goals_b in range(max_goals + 1):
            prob = (
                poisson_prob(goals_a, expected_goals_a)
                *
                poisson_prob(goals_b, expected_goals_b)
            )

            rows.append(
                {
                    "team_a": team_a,
                    "team_b": team_b,
                    "team_a_display": team_a_display,
                    "team_b_display": team_b_display,
                    "score": f"{goals_a}-{goals_b}",
                    "goals_a": goals_a,
                    "goals_b": goals_b,
                    "score_probability": prob,
                    "expected_goals_a": expected_goals_a,
                    "expected_goals_b": expected_goals_b
                }
            )

    df = pd.DataFrame(rows)

    # Normalizar para que las probabilidades de los scores sumen 100%
    df["score_probability"] = df["score_probability"] / df["score_probability"].sum()

    win_a_probability = df.loc[df["goals_a"] > df["goals_b"], "score_probability"].sum()
    draw_probability = df.loc[df["goals_a"] == df["goals_b"], "score_probability"].sum()
    win_b_probability = df.loc[df["goals_a"] < df["goals_b"], "score_probability"].sum()

    df["win_a_probability"] = win_a_probability
    df["draw_probability"] = draw_probability
    df["win_b_probability"] = win_b_probability

    return df

# Generar todos los cruces posibles
all_matches = []

for i, team_a_row in teams.iterrows():
    for j, team_b_row in teams.iterrows():
        if team_a_row["team"] != team_b_row["team"]:
            match_df = simulate_match(team_a_row, team_b_row)
            all_matches.append(match_df)

world_cup_match_simulator = pd.concat(all_matches, ignore_index=True)

# Ordenar
world_cup_match_simulator = world_cup_match_simulator.sort_values(
    ["team_a", "team_b", "score_probability"],
    ascending=[True, True, False]
).reset_index(drop=True)

# Redondear
round_cols = [
    "score_probability",
    "win_a_probability",
    "draw_probability",
    "win_b_probability",
    "expected_goals_a",
    "expected_goals_b"
]

for col in round_cols:
    world_cup_match_simulator[col] = world_cup_match_simulator[col].round(6)

print("Filas generadas:", len(world_cup_match_simulator))
print("Equipos en team_a:", world_cup_match_simulator["team_a"].nunique())
print("Equipos en team_b:", world_cup_match_simulator["team_b"].nunique())

# Buscar archivos existentes del simulador
simulator_files = []

for root in search_roots:
    if root.exists():
        simulator_files.extend(list(root.rglob("world_cup_match_simulator.xlsx")))

simulator_files = list(dict.fromkeys(simulator_files))

# Si no encuentra, guardar al lado del ranking
if not simulator_files:
    simulator_files = [ranking_file.parent / "world_cup_match_simulator.xlsx"]

# Exportar reemplazando todos los simuladores encontrados
for file in simulator_files:
    try:
        world_cup_match_simulator.to_excel(file, index=False)
        print("Simulador exportado en:")
        print(file)
    except PermissionError:
        print("ERROR: Cerrá el Excel abierto y volvé a correr. No se pudo escribir:")
        print(file)

display(
    world_cup_match_simulator[
        [
            "team_a_display",
            "team_b_display",
            "score",
            "score_probability",
            "win_a_probability",
            "draw_probability",
            "win_b_probability",
            "expected_goals_a",
            "expected_goals_b"
        ]
    ].head(20)
)
```

```python
import pandas as pd
import numpy as np
from pathlib import Path

print("Generando simulación del torneo...")

N_SIMULATIONS = 25000
RANDOM_SEED = 42
rng = np.random.default_rng(RANDOM_SEED)

# Buscar archivos base
search_roots = [
    Path.cwd(),
    Path.cwd().parent,
    Path.cwd().parent.parent
]

ranking_files = []
simulator_files = []

for root in search_roots:
    if root.exists():
        ranking_files.extend(list(root.rglob("world_cup_live_power_ranking.xlsx")))
        simulator_files.extend(list(root.rglob("world_cup_match_simulator.xlsx")))

ranking_files = list(dict.fromkeys(ranking_files))
simulator_files = list(dict.fromkeys(simulator_files))

if not ranking_files:
    raise FileNotFoundError("No encontré world_cup_live_power_ranking.xlsx")

if not simulator_files:
    raise FileNotFoundError("No encontré world_cup_match_simulator.xlsx")

ranking_file = ranking_files[0]
simulator_file = simulator_files[0]

print("Ranking:", ranking_file)
print("Simulador:", simulator_file)

ranking = pd.read_excel(ranking_file)
simulator = pd.read_excel(simulator_file)

# Normalizar probabilidades por si vienen en escala 0-100
for col in ["win_a_probability", "draw_probability", "win_b_probability", "score_probability"]:
    simulator[col] = pd.to_numeric(simulator[col], errors="coerce")
    if simulator[col].max() > 1:
        simulator[col] = simulator[col] / 100

# Usamos Top 16 del ranking para la ruta visual del torneo
top16 = ranking.sort_values("power_score_live", ascending=False).head(16).copy()
top16 = top16.reset_index(drop=True)
top16["seed"] = range(1, 17)

print("Equipos usados en la ruta proyectada:", top16["team"].nunique())

team_info = ranking.set_index("team").to_dict("index")

# Probabilidad de avanzar en cruce directo
def get_advance_probability(team_a, team_b):
    row = simulator[
        (simulator["team_a"] == team_a) &
        (simulator["team_b"] == team_b)
    ]

    if row.empty:
        row = simulator[
            (simulator["team_a"] == team_b) &
            (simulator["team_b"] == team_a)
        ]

        if row.empty:
            return 0.5

        win_a = float(row["win_a_probability"].iloc[0])
        draw = float(row["draw_probability"].iloc[0])
        win_b = float(row["win_b_probability"].iloc[0])

        # Si el partido está invertido, la probabilidad de team_a es win_b + mitad empate
        return win_b + draw * 0.5

    win_a = float(row["win_a_probability"].iloc[0])
    draw = float(row["draw_probability"].iloc[0])

    return win_a + draw * 0.5

def play_match(team_a, team_b):
    p_a = get_advance_probability(team_a, team_b)
    return team_a if rng.random() < p_a else team_b

# Orden de bracket visual: 1 vs 16, 8 vs 9, 4 vs 13, etc.
seed_pairs = [
    (1, 16),
    (8, 9),
    (4, 13),
    (5, 12),
    (2, 15),
    (7, 10),
    (3, 14),
    (6, 11)
]

seed_to_team = dict(zip(top16["seed"], top16["team"]))

initial_matches = [
    (seed_to_team[a], seed_to_team[b])
    for a, b in seed_pairs
]

# Contadores de rondas
teams_all = ranking["team"].unique()

round_counts = {
    team: {
        "round_16_probability": 0,
        "quarterfinal_probability": 0,
        "semifinal_probability": 0,
        "final_probability": 0,
        "champion_probability": 0
    }
    for team in teams_all
}

# Simulaciones Monte Carlo
for _ in range(N_SIMULATIONS):
    # Participan en ruta visual top 16
    for team in top16["team"]:
        round_counts[team]["round_16_probability"] += 1

    # Ronda 1
    winners_1 = []
    for a, b in initial_matches:
        winner = play_match(a, b)
        winners_1.append(winner)
        round_counts[winner]["quarterfinal_probability"] += 1

    # Cuartos
    matches_2 = [
        (winners_1[0], winners_1[1]),
        (winners_1[2], winners_1[3]),
        (winners_1[4], winners_1[5]),
        (winners_1[6], winners_1[7])
    ]

    winners_2 = []
    for a, b in matches_2:
        winner = play_match(a, b)
        winners_2.append(winner)
        round_counts[winner]["semifinal_probability"] += 1

    # Semifinales
    matches_3 = [
        (winners_2[0], winners_2[1]),
        (winners_2[2], winners_2[3])
    ]

    winners_3 = []
    for a, b in matches_3:
        winner = play_match(a, b)
        winners_3.append(winner)
        round_counts[winner]["final_probability"] += 1

    # Final
    champion = play_match(winners_3[0], winners_3[1])
    round_counts[champion]["champion_probability"] += 1

# Crear tabla de probabilidades por ronda
round_probabilities = []

for team, counts in round_counts.items():
    info = team_info.get(team, {})
    row = {
        "team": team,
        "team_display": info.get("team_display", team),
        "flag": info.get("flag", ""),
        "confederation": info.get("confederation", ""),
        "ranking": info.get("ranking", np.nan),
        "power_score_live": info.get("power_score_live", np.nan),
        "rating": info.get("rating", np.nan),
    }

    for col, value in counts.items():
        row[col] = value / N_SIMULATIONS

    round_probabilities.append(row)

round_probabilities = pd.DataFrame(round_probabilities)

round_probabilities = round_probabilities.sort_values(
    "champion_probability",
    ascending=False
).reset_index(drop=True)

# Top candidatos
top_candidates = round_probabilities.head(8).copy()

# Campeón proyectado
champion_row = round_probabilities.iloc[0]
champion_team = champion_row["team"]

# Probabilidades por confederación
confederation_probabilities = (
    round_probabilities
    .groupby("confederation", as_index=False)["champion_probability"]
    .sum()
    .sort_values("champion_probability", ascending=False)
)

# Equipos sorpresa: equipos fuera del top 8 del ranking inicial con mejor probabilidad de campeón
surprises = (
    round_probabilities[
        round_probabilities["ranking"] > 8
    ]
    .sort_values("champion_probability", ascending=False)
    .head(5)
    .copy()
)

# Progreso del campeón por ronda
round_progress_champion = pd.DataFrame([
    {
        "round": "16 Avos",
        "probability": champion_row["round_16_probability"]
    },
    {
        "round": "Cuartos",
        "probability": champion_row["quarterfinal_probability"]
    },
    {
        "round": "Semifinales",
        "probability": champion_row["semifinal_probability"]
    },
    {
        "round": "Final",
        "probability": champion_row["final_probability"]
    },
    {
        "round": "Campeón",
        "probability": champion_row["champion_probability"]
    }
])

# Bracket determinístico: siempre avanza el equipo con mayor probabilidad
def deterministic_winner(team_a, team_b):
    p_a = get_advance_probability(team_a, team_b)
    if p_a >= 0.5:
        return team_a, p_a
    else:
        return team_b, 1 - p_a

bracket_rows = []

def display_team(team):
    return team_info.get(team, {}).get("team_display", team)

# Ronda inicial
r1_winners = []
for idx, (a, b) in enumerate(initial_matches, start=1):
    winner, p = deterministic_winner(a, b)
    r1_winners.append(winner)

    bracket_rows.append({
        "stage": "16 Avos",
        "match": idx,
        "team_a": a,
        "team_b": b,
        "team_a_display": display_team(a),
        "team_b_display": display_team(b),
        "winner": winner,
        "winner_display": display_team(winner),
        "winner_probability": p
    })

# Cuartos
r2_matches = [
    (r1_winners[0], r1_winners[1]),
    (r1_winners[2], r1_winners[3]),
    (r1_winners[4], r1_winners[5]),
    (r1_winners[6], r1_winners[7])
]

r2_winners = []
for idx, (a, b) in enumerate(r2_matches, start=1):
    winner, p = deterministic_winner(a, b)
    r2_winners.append(winner)

    bracket_rows.append({
        "stage": "Cuartos",
        "match": idx,
        "team_a": a,
        "team_b": b,
        "team_a_display": display_team(a),
        "team_b_display": display_team(b),
        "winner": winner,
        "winner_display": display_team(winner),
        "winner_probability": p
    })

# Semifinales
r3_matches = [
    (r2_winners[0], r2_winners[1]),
    (r2_winners[2], r2_winners[3])
]

r3_winners = []
semifinal_rows = []

for idx, (a, b) in enumerate(r3_matches, start=1):
    winner, p = deterministic_winner(a, b)
    r3_winners.append(winner)

    row = {
        "stage": "Semifinales",
        "match": idx,
        "team_a": a,
        "team_b": b,
        "team_a_display": display_team(a),
        "team_b_display": display_team(b),
        "winner": winner,
        "winner_display": display_team(winner),
        "winner_probability": p
    }

    bracket_rows.append(row)
    semifinal_rows.append(row)

# Final
final_winner, final_p = deterministic_winner(r3_winners[0], r3_winners[1])

final_row = {
    "stage": "Final",
    "match": 1,
    "team_a": r3_winners[0],
    "team_b": r3_winners[1],
    "team_a_display": display_team(r3_winners[0]),
    "team_b_display": display_team(r3_winners[1]),
    "winner": final_winner,
    "winner_display": display_team(final_winner),
    "winner_probability": final_p
}

bracket_rows.append(final_row)

most_likely_bracket = pd.DataFrame(bracket_rows)

# Tabla resumen de tarjetas
key_cards = pd.DataFrame([{
    "champion": champion_team,
    "champion_display": champion_row["team_display"],
    "champion_probability": champion_row["champion_probability"],
    "final_probability": champion_row["final_probability"],
    "semifinal_probability": champion_row["semifinal_probability"],
    "teams_simulated": ranking["team"].nunique(),
    "simulations": N_SIMULATIONS,
    "final_team_a_display": final_row["team_a_display"],
    "final_team_b_display": final_row["team_b_display"],
    "final_match_probability": final_row["winner_probability"],
    "semi_1_team_a_display": semifinal_rows[0]["team_a_display"],
    "semi_1_team_b_display": semifinal_rows[0]["team_b_display"],
    "semi_1_probability": semifinal_rows[0]["winner_probability"],
    "semi_2_team_a_display": semifinal_rows[1]["team_a_display"],
    "semi_2_team_b_display": semifinal_rows[1]["team_b_display"],
    "semi_2_probability": semifinal_rows[1]["winner_probability"],
}])

# Exportar a Excel con varias hojas
output_file = ranking_file.parent / "world_cup_tournament_simulation.xlsx"

with pd.ExcelWriter(output_file, engine="openpyxl") as writer:
    key_cards.to_excel(writer, sheet_name="key_cards", index=False)
    round_probabilities.to_excel(writer, sheet_name="round_probabilities", index=False)
    top_candidates.to_excel(writer, sheet_name="top_candidates", index=False)
    round_progress_champion.to_excel(writer, sheet_name="round_progress_champion", index=False)
    confederation_probabilities.to_excel(writer, sheet_name="confederation_probabilities", index=False)
    surprises.to_excel(writer, sheet_name="surprises", index=False)
    most_likely_bracket.to_excel(writer, sheet_name="most_likely_bracket", index=False)

print("Archivo creado correctamente:")
print(output_file)

print("\nCampeón proyectado:")
display(key_cards)

print("\nTop candidatos:")
display(top_candidates[[
    "team_display",
    "champion_probability",
    "final_probability",
    "semifinal_probability",
    "confederation"
]])

print("\nBracket más probable:")
display(most_likely_bracket)
```

```python
import pandas as pd
import numpy as np
from pathlib import Path

print("Generando simulación rápida del torneo...")

N_SIMULATIONS = 25000
RANDOM_SEED = 42
rng = np.random.default_rng(RANDOM_SEED)

# Buscar archivos
search_roots = [
    Path.cwd(),
    Path.cwd().parent,
    Path.cwd().parent.parent
]

ranking_files = []
simulator_files = []

for root in search_roots:
    if root.exists():
        ranking_files.extend(list(root.rglob("world_cup_live_power_ranking.xlsx")))
        simulator_files.extend(list(root.rglob("world_cup_match_simulator.xlsx")))

ranking_files = list(dict.fromkeys(ranking_files))
simulator_files = list(dict.fromkeys(simulator_files))

if not ranking_files:
    raise FileNotFoundError("No encontré world_cup_live_power_ranking.xlsx")

if not simulator_files:
    raise FileNotFoundError("No encontré world_cup_match_simulator.xlsx")

ranking_file = ranking_files[0]
simulator_file = simulator_files[0]

print("Ranking:", ranking_file)
print("Simulador:", simulator_file)

ranking = pd.read_excel(ranking_file)
simulator = pd.read_excel(simulator_file)

# Normalizar probabilidades
for col in ["win_a_probability", "draw_probability", "win_b_probability", "score_probability"]:
    simulator[col] = pd.to_numeric(simulator[col], errors="coerce")
    if simulator[col].max() > 1:
        simulator[col] = simulator[col] / 100

# Tomamos Top 16 para la ruta proyectada del torneo
top16 = ranking.sort_values("power_score_live", ascending=False).head(16).copy()
top16 = top16.reset_index(drop=True)
top16["seed"] = range(1, 17)

print("Equipos usados en bracket:", top16["team"].nunique())

team_info = ranking.set_index("team").to_dict("index")

# Crear diccionario rápido de probabilidades de avance
pair_probs = (
    simulator[
        ["team_a", "team_b", "win_a_probability", "draw_probability", "win_b_probability"]
    ]
    .drop_duplicates(subset=["team_a", "team_b"])
    .copy()
)

advance_prob = {}

for _, row in pair_probs.iterrows():
    a = row["team_a"]
    b = row["team_b"]
    win_a = float(row["win_a_probability"])
    draw = float(row["draw_probability"])
    win_b = float(row["win_b_probability"])
    
    advance_prob[(a, b)] = win_a + draw * 0.5
    advance_prob[(b, a)] = win_b + draw * 0.5

def get_advance_probability(team_a, team_b):
    return advance_prob.get((team_a, team_b), 0.5)

def play_match(team_a, team_b):
    p_a = get_advance_probability(team_a, team_b)
    return team_a if rng.random() < p_a else team_b

def deterministic_winner(team_a, team_b):
    p_a = get_advance_probability(team_a, team_b)
    if p_a >= 0.5:
        return team_a, p_a
    else:
        return team_b, 1 - p_a

def display_team(team):
    return team_info.get(team, {}).get("team_display", team)

# Bracket: 1 vs 16, 8 vs 9, etc.
seed_pairs = [
    (1, 16),
    (8, 9),
    (4, 13),
    (5, 12),
    (2, 15),
    (7, 10),
    (3, 14),
    (6, 11)
]

seed_to_team = dict(zip(top16["seed"], top16["team"]))

initial_matches = [
    (seed_to_team[a], seed_to_team[b])
    for a, b in seed_pairs
]

# Contadores solo para equipos del top16
round_counts = {
    team: {
        "round_16_probability": 0,
        "quarterfinal_probability": 0,
        "semifinal_probability": 0,
        "final_probability": 0,
        "champion_probability": 0
    }
    for team in ranking["team"].unique()
}

# Simulaciones
for _ in range(N_SIMULATIONS):

    for team in top16["team"]:
        round_counts[team]["round_16_probability"] += 1

    winners_1 = []
    for a, b in initial_matches:
        winner = play_match(a, b)
        winners_1.append(winner)
        round_counts[winner]["quarterfinal_probability"] += 1

    matches_2 = [
        (winners_1[0], winners_1[1]),
        (winners_1[2], winners_1[3]),
        (winners_1[4], winners_1[5]),
        (winners_1[6], winners_1[7])
    ]

    winners_2 = []
    for a, b in matches_2:
        winner = play_match(a, b)
        winners_2.append(winner)
        round_counts[winner]["semifinal_probability"] += 1

    matches_3 = [
        (winners_2[0], winners_2[1]),
        (winners_2[2], winners_2[3])
    ]

    winners_3 = []
    for a, b in matches_3:
        winner = play_match(a, b)
        winners_3.append(winner)
        round_counts[winner]["final_probability"] += 1

    champion = play_match(winners_3[0], winners_3[1])
    round_counts[champion]["champion_probability"] += 1

# Tabla probabilidades por ronda
round_probabilities = []

for team, counts in round_counts.items():
    info = team_info.get(team, {})
    row = {
        "team": team,
        "team_display": info.get("team_display", team),
        "flag": info.get("flag", ""),
        "confederation": info.get("confederation", ""),
        "ranking": info.get("ranking", np.nan),
        "power_score_live": info.get("power_score_live", np.nan),
        "rating": info.get("rating", np.nan),
    }

    for col, value in counts.items():
        row[col] = value / N_SIMULATIONS

    round_probabilities.append(row)

round_probabilities = pd.DataFrame(round_probabilities)

round_probabilities = round_probabilities.sort_values(
    "champion_probability",
    ascending=False
).reset_index(drop=True)

top_candidates = round_probabilities.head(8).copy()

champion_row = round_probabilities.iloc[0]
champion_team = champion_row["team"]

confederation_probabilities = (
    round_probabilities
    .groupby("confederation", as_index=False)["champion_probability"]
    .sum()
    .sort_values("champion_probability", ascending=False)
)

surprises = (
    round_probabilities[
        round_probabilities["ranking"] > 8
    ]
    .sort_values("champion_probability", ascending=False)
    .head(5)
    .copy()
)

round_progress_champion = pd.DataFrame([
    {"round": "16 Avos", "probability": champion_row["round_16_probability"]},
    {"round": "Cuartos", "probability": champion_row["quarterfinal_probability"]},
    {"round": "Semifinales", "probability": champion_row["semifinal_probability"]},
    {"round": "Final", "probability": champion_row["final_probability"]},
    {"round": "Campeón", "probability": champion_row["champion_probability"]},
])

# Bracket determinístico para llenar el mockup
bracket_rows = []

r1_winners = []
for idx, (a, b) in enumerate(initial_matches, start=1):
    winner, p = deterministic_winner(a, b)
    r1_winners.append(winner)
    bracket_rows.append({
        "stage": "16 Avos",
        "match": idx,
        "team_a": a,
        "team_b": b,
        "team_a_display": display_team(a),
        "team_b_display": display_team(b),
        "winner": winner,
        "winner_display": display_team(winner),
        "winner_probability": p
    })

r2_matches = [
    (r1_winners[0], r1_winners[1]),
    (r1_winners[2], r1_winners[3]),
    (r1_winners[4], r1_winners[5]),
    (r1_winners[6], r1_winners[7])
]

r2_winners = []
for idx, (a, b) in enumerate(r2_matches, start=1):
    winner, p = deterministic_winner(a, b)
    r2_winners.append(winner)
    bracket_rows.append({
        "stage": "Cuartos",
        "match": idx,
        "team_a": a,
        "team_b": b,
        "team_a_display": display_team(a),
        "team_b_display": display_team(b),
        "winner": winner,
        "winner_display": display_team(winner),
        "winner_probability": p
    })

r3_matches = [
    (r2_winners[0], r2_winners[1]),
    (r2_winners[2], r2_winners[3])
]

r3_winners = []
semifinal_rows = []

for idx, (a, b) in enumerate(r3_matches, start=1):
    winner, p = deterministic_winner(a, b)
    r3_winners.append(winner)

    row = {
        "stage": "Semifinales",
        "match": idx,
        "team_a": a,
        "team_b": b,
        "team_a_display": display_team(a),
        "team_b_display": display_team(b),
        "winner": winner,
        "winner_display": display_team(winner),
        "winner_probability": p
    }

    bracket_rows.append(row)
    semifinal_rows.append(row)

final_winner, final_p = deterministic_winner(r3_winners[0], r3_winners[1])

final_row = {
    "stage": "Final",
    "match": 1,
    "team_a": r3_winners[0],
    "team_b": r3_winners[1],
    "team_a_display": display_team(r3_winners[0]),
    "team_b_display": display_team(r3_winners[1]),
    "winner": final_winner,
    "winner_display": display_team(final_winner),
    "winner_probability": final_p
}

bracket_rows.append(final_row)

most_likely_bracket = pd.DataFrame(bracket_rows)

key_cards = pd.DataFrame([{
    "champion": champion_team,
    "champion_display": champion_row["team_display"],
    "champion_probability": champion_row["champion_probability"],
    "final_probability": champion_row["final_probability"],
    "semifinal_probability": champion_row["semifinal_probability"],
    "teams_simulated": ranking["team"].nunique(),
    "simulations": N_SIMULATIONS,
    "final_team_a_display": final_row["team_a_display"],
    "final_team_b_display": final_row["team_b_display"],
    "final_match_probability": final_row["winner_probability"],
    "semi_1_team_a_display": semifinal_rows[0]["team_a_display"],
    "semi_1_team_b_display": semifinal_rows[0]["team_b_display"],
    "semi_1_probability": semifinal_rows[0]["winner_probability"],
    "semi_2_team_a_display": semifinal_rows[1]["team_a_display"],
    "semi_2_team_b_display": semifinal_rows[1]["team_b_display"],
    "semi_2_probability": semifinal_rows[1]["winner_probability"],
}])

output_file = ranking_file.parent / "world_cup_tournament_simulation.xlsx"

with pd.ExcelWriter(output_file, engine="openpyxl") as writer:
    key_cards.to_excel(writer, sheet_name="key_cards", index=False)
    round_probabilities.to_excel(writer, sheet_name="round_probabilities", index=False)
    top_candidates.to_excel(writer, sheet_name="top_candidates", index=False)
    round_progress_champion.to_excel(writer, sheet_name="round_progress_champion", index=False)
    confederation_probabilities.to_excel(writer, sheet_name="confederation_probabilities", index=False)
    surprises.to_excel(writer, sheet_name="surprises", index=False)
    most_likely_bracket.to_excel(writer, sheet_name="most_likely_bracket", index=False)

print("Archivo creado correctamente:")
print(output_file)

print("Simulaciones:", N_SIMULATIONS)
print("Campeón proyectado:", champion_row["team_display"])
print("Prob campeón:", round(champion_row["champion_probability"] * 100, 1), "%")

display(key_cards)
display(top_candidates[["team_display", "champion_probability", "final_probability", "semifinal_probability"]])
display(most_likely_bracket)
```
