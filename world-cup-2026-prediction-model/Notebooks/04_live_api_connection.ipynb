```python
def open_api_get(endpoint):
    url = f"{OPEN_BASE_URL}{endpoint}"
    
    response = requests.get(url, timeout=30)
    
    print("Endpoint:", endpoint)
    print("Status code:", response.status_code)
    
    if response.status_code != 200:
        print(response.text)
        raise Exception(f"Error {response.status_code}")
    
    return response.json()


def unwrap_api_response(data):
    """
    Esta función intenta encontrar la lista principal dentro de la respuesta,
    aunque la API cambie un poco el nombre del campo.
    """
    
    if isinstance(data, list):
        return data
    
    if isinstance(data, dict):
        possible_keys = [
            "data",
            "response",
            "teams",
            "games",
            "matches",
            "groups",
            "stadiums",
            "result"
        ]
        
        for key in possible_keys:
            if key in data:
                return data[key]
    
    return data
```

```python
import pandas as pd
import numpy as np
from pathlib import Path

print("Creando tabla normalizada de partidos reales...")

live_file = OUTPUT_DIR / "worldcup_open_live_data.xlsx"

sheets = pd.read_excel(live_file, sheet_name=None)

games = sheets["games"].copy()
teams = sheets["teams"].copy()
groups = sheets["groups"].copy()

# Normalizar nombres de columnas
games.columns = [str(c).strip() for c in games.columns]

print("Filas en games:", len(games))
print("Columnas disponibles:")
print(games.columns.tolist())

def find_col(df, keywords, required=False):
    """
    Busca una columna que contenga todas o algunas palabras clave.
    """
    cols = df.columns.tolist()
    cols_lower = {c: c.lower() for c in cols}
    
    # Primero busca coincidencias fuertes
    for col, low in cols_lower.items():
        if all(k.lower() in low for k in keywords):
            return col
    
    # Después busca coincidencias parciales
    for col, low in cols_lower.items():
        if any(k.lower() in low for k in keywords):
            return col
    
    if required:
        raise ValueError(f"No encontré columna para keywords: {keywords}")
    
    return None

# Detectar columnas probables
id_col = find_col(games, ["id"])
date_col = find_col(games, ["date"])
round_col = find_col(games, ["round"]) or find_col(games, ["stage"]) or find_col(games, ["phase"])
status_col = find_col(games, ["status"])

home_team_col = (
    find_col(games, ["home", "name"]) or
    find_col(games, ["home_team"]) or
    find_col(games, ["home"])
)

away_team_col = (
    find_col(games, ["away", "name"]) or
    find_col(games, ["away_team"]) or
    find_col(games, ["away"])
)

home_score_col = (
    find_col(games, ["home", "score"]) or
    find_col(games, ["home", "goal"]) or
    find_col(games, ["score1"])
)

away_score_col = (
    find_col(games, ["away", "score"]) or
    find_col(games, ["away", "goal"]) or
    find_col(games, ["score2"])
)

group_col = find_col(games, ["group"])

print("\nColumnas detectadas:")
print("id_col:", id_col)
print("date_col:", date_col)
print("round_col:", round_col)
print("status_col:", status_col)
print("home_team_col:", home_team_col)
print("away_team_col:", away_team_col)
print("home_score_col:", home_score_col)
print("away_score_col:", away_score_col)
print("group_col:", group_col)

# Crear tabla normalizada
official_games = pd.DataFrame()

official_games["match_id"] = games[id_col] if id_col else range(1, len(games) + 1)
official_games["date"] = games[date_col] if date_col else None
official_games["round"] = games[round_col] if round_col else None
official_games["status"] = games[status_col] if status_col else None
official_games["group"] = games[group_col] if group_col else None
official_games["team_a"] = games[home_team_col] if home_team_col else None
official_games["team_b"] = games[away_team_col] if away_team_col else None
official_games["goals_a"] = games[home_score_col] if home_score_col else None
official_games["goals_b"] = games[away_score_col] if away_score_col else None

# Limpiar textos
for col in ["round", "status", "group", "team_a", "team_b"]:
    official_games[col] = official_games[col].astype(str).replace("nan", "")

# Convertir goles
official_games["goals_a"] = pd.to_numeric(official_games["goals_a"], errors="coerce")
official_games["goals_b"] = pd.to_numeric(official_games["goals_b"], errors="coerce")

# Detectar partidos terminados
official_games["is_finished"] = official_games["status"].str.contains(
    "finished|complete|full|ft|final",
    case=False,
    na=False
)

# Determinar ganador real si hay resultado
def get_real_winner(row):
    if pd.isna(row["goals_a"]) or pd.isna(row["goals_b"]):
        return ""
    
    if row["goals_a"] > row["goals_b"]:
        return row["team_a"]
    elif row["goals_b"] > row["goals_a"]:
        return row["team_b"]
    else:
        return "Draw"

official_games["real_winner"] = official_games.apply(get_real_winner, axis=1)

# Clasificar fase
official_games["is_group_stage"] = official_games["round"].str.contains(
    "group|grupo",
    case=False,
    na=False
)

official_games["is_knockout"] = ~official_games["is_group_stage"]

# Ordenar
official_games = official_games.sort_values(["date", "match_id"], na_position="last").reset_index(drop=True)

# Exportar
output_file = OUTPUT_DIR / "official_games_normalized.xlsx"
official_games.to_excel(output_file, index=False)

print("\nArchivo creado:")
print(output_file)

print("\nResumen:")
print("Partidos totales:", len(official_games))
print("Fase de grupos:", official_games["is_group_stage"].sum())
print("Eliminación directa:", official_games["is_knockout"].sum())
print("Terminados:", official_games["is_finished"].sum())

print("\nRondas detectadas:")
display(
    official_games["round"]
    .value_counts()
    .reset_index()
    .rename(columns={"index": "round", "round": "matches"})
)

display(official_games.head(30))
```

```python
import pandas as pd
import numpy as np
import re
import unicodedata
from pathlib import Path

print("Generando simulación sobre bracket real del Mundial 2026...")

N_SIMULATIONS = 25000
RANDOM_SEED = 42
rng = np.random.default_rng(RANDOM_SEED)

# =========================
# 1. Buscar archivos
# =========================

search_roots = [
    Path.cwd(),
    Path.cwd().parent,
    Path.cwd().parent.parent
]

bracket_files = []
ranking_files = []
simulator_files = []

for root in search_roots:
    if root.exists():
        bracket_files.extend(list(root.rglob("official_tournament_bracket_normalized.xlsx")))
        ranking_files.extend(list(root.rglob("world_cup_live_power_ranking.xlsx")))
        simulator_files.extend(list(root.rglob("world_cup_match_simulator.xlsx")))

bracket_files = list(dict.fromkeys(bracket_files))
ranking_files = list(dict.fromkeys(ranking_files))
simulator_files = list(dict.fromkeys(simulator_files))

if not bracket_files:
    raise FileNotFoundError("No encontré official_tournament_bracket_normalized.xlsx")

if not ranking_files:
    raise FileNotFoundError("No encontré world_cup_live_power_ranking.xlsx")

if not simulator_files:
    raise FileNotFoundError("No encontré world_cup_match_simulator.xlsx")

bracket_file = bracket_files[0]
ranking_file = ranking_files[0]
simulator_file = simulator_files[0]

print("Bracket real:", bracket_file)
print("Ranking:", ranking_file)
print("Simulador:", simulator_file)

# =========================
# 2. Cargar datos
# =========================

bracket_sheets = pd.read_excel(bracket_file, sheet_name=None)

knockout = bracket_sheets["knockout_bracket"].copy()
teams_api = bracket_sheets["teams"].copy()
ranking = pd.read_excel(ranking_file)
simulator = pd.read_excel(simulator_file)

# =========================
# 3. Normalizar nombres
# =========================

name_aliases = {
    "Curaçao": "Curacao",
    "Curacao": "Curacao",
    "Czechia": "Czech Republic",
    "Czech Republic": "Czech Republic",
    "Democratic Republic of the Congo": "Congo",
    "DR Congo": "Congo",
    "Congo DR": "Congo",
    "Congo": "Congo",
    "United States": "United States",
    "USA": "United States",
}

def normalize_text(value):
    if pd.isna(value):
        return ""
    value = str(value).strip()
    value = unicodedata.normalize("NFKD", value)
    value = "".join([c for c in value if not unicodedata.combining(c)])
    return value

def canonical_team(value):
    if pd.isna(value):
        return ""
    
    value = str(value).strip()
    
    if value in name_aliases:
        return name_aliases[value]
    
    value_clean = normalize_text(value)
    
    for original, mapped in name_aliases.items():
        if normalize_text(original).lower() == value_clean.lower():
            return mapped
    
    return value

for col in ["team_a", "team_b", "real_winner"]:
    if col in knockout.columns:
        knockout[col] = knockout[col].apply(canonical_team)

ranking["team"] = ranking["team"].apply(canonical_team)

if "team_display" not in ranking.columns:
    ranking["team_display"] = ranking["team"]

ranking = ranking.drop_duplicates(subset=["team"]).reset_index(drop=True)

team_info = ranking.set_index("team").to_dict("index")

def team_display(team):
    team = canonical_team(team)
    return team_info.get(team, {}).get("team_display", team)

def team_confederation(team):
    team = canonical_team(team)
    return team_info.get(team, {}).get("confederation", "")

def team_ranking(team):
    team = canonical_team(team)
    return team_info.get(team, {}).get("ranking", 99)

def team_power(team):
    team = canonical_team(team)
    return float(team_info.get(team, {}).get("power_score_live", 0.50))

def team_rating(team):
    team = canonical_team(team)
    return float(team_info.get(team, {}).get("rating", 1500))

# =========================
# 4. Resolver terceros pendientes
# =========================

def parse_candidates(text):
    if pd.isna(text) or str(text).strip() == "":
        return []
    
    text = str(text)
    
    matches = re.findall(r"([A-L]\d)\s+([^,]+)", text)
    
    candidates = []
    
    for slot, team in matches:
        candidates.append({
            "slot": slot,
            "team": canonical_team(team.strip())
        })
    
    return candidates

pending_slots = []

for idx, row in knockout.iterrows():
    if row["round_type"] != "r32":
        continue
    
    for side in ["a", "b"]:
        status_col = f"team_{side}_status"
        candidates_col = f"team_{side}_andidates"
        
        # Por si la columna está bien escrita como candidates
        candidates_col = f"team_{side}_candidates"
        
        if status_col in knockout.columns and candidates_col in knockout.columns:
            if str(row[status_col]) == "pending_third_place_assignment":
                candidates = parse_candidates(row[candidates_col])
                pending_slots.append({
                    "row_index": idx,
                    "side": side,
                    "match_id": row["match_id"],
                    "candidates": candidates
                })

# Terceros ya asignados oficialmente
used_third_place_teams = set()

for _, row in knockout[knockout["round_type"] == "r32"].iterrows():
    for side in ["a", "b"]:
        raw_col = f"team_{side}_raw_label"
        team_col = f"team_{side}"
        status_col = f"team_{side}_status"
        
        raw_label = str(row.get(raw_col, ""))
        status = str(row.get(status_col, ""))
        team = canonical_team(row.get(team_col, ""))
        
        if "3rd Group" in raw_label and status != "pending_third_place_assignment" and team:
            used_third_place_teams.add(team)

third_place_assignments = []

# Resolver primero los slots con menos candidatos
pending_slots = sorted(pending_slots, key=lambda x: len(x["candidates"]))

for slot in pending_slots:
    available = [
        c["team"]
        for c in slot["candidates"]
        if c["team"] not in used_third_place_teams
    ]
    
    if not available:
        available = [c["team"] for c in slot["candidates"]]
    
    # Elegimos el candidato más fuerte por Power Score
    selected = sorted(
        available,
        key=lambda t: team_power(t),
        reverse=True
    )[0]
    
    used_third_place_teams.add(selected)
    
    team_col = f"team_{slot['side']}"
    status_col = f"team_{slot['side']}_status"
    
    knockout.loc[slot["row_index"], team_col] = selected
    knockout.loc[slot["row_index"], status_col] = "resolved_third_place_model"
    
    third_place_assignments.append({
        "match_id": slot["match_id"],
        "side": slot["side"],
        "selected_team": selected,
        "selected_team_display": team_display(selected),
        "candidate_options": ", ".join([c["team"] for c in slot["candidates"]])
    })

third_place_assignments = pd.DataFrame(third_place_assignments)

print("Terceros pendientes resueltos:")
display(third_place_assignments)

# =========================
# 5. Probabilidades de partidos
# =========================

for col in ["team_a", "team_b"]:
    simulator[col] = simulator[col].apply(canonical_team)

for col in ["win_a_probability", "draw_probability", "win_b_probability"]:
    simulator[col] = pd.to_numeric(simulator[col], errors="coerce")
    if simulator[col].max() > 1:
        simulator[col] = simulator[col] / 100

pair_probs = (
    simulator[
        ["team_a", "team_b", "win_a_probability", "draw_probability", "win_b_probability"]
    ]
    .drop_duplicates(subset=["team_a", "team_b"])
    .copy()
)

advance_prob = {}

for _, row in pair_probs.iterrows():
    a = row["team_a"]
    b = row["team_b"]
    win_a = float(row["win_a_probability"])
    draw = float(row["draw_probability"])
    win_b = float(row["win_b_probability"])
    
    advance_prob[(a, b)] = win_a + draw * 0.5
    advance_prob[(b, a)] = win_b + draw * 0.5

def fallback_probability(team_a, team_b):
    power_diff = team_power(team_a) - team_power(team_b)
    rating_diff = (team_rating(team_a) - team_rating(team_b)) / 600
    
    x = power_diff * 2.5 + rating_diff
    p = 1 / (1 + np.exp(-x))
    
    return float(np.clip(p, 0.05, 0.95))

def get_advance_probability(team_a, team_b):
    team_a = canonical_team(team_a)
    team_b = canonical_team(team_b)
    
    if team_a == team_b:
        return 1.0
    
    if (team_a, team_b) in advance_prob:
        return advance_prob[(team_a, team_b)]
    
    return fallback_probability(team_a, team_b)

# =========================
# 6. Resolver participantes
# =========================

knockout["match_id"] = pd.to_numeric(knockout["match_id"], errors="coerce").astype(int)

match_table = knockout.set_index("match_id").to_dict("index")

def resolve_participant(label, winners, losers=None):
    label = str(label)
    
    if label.startswith("Winner Match"):
        match_id = int(re.findall(r"\d+", label)[0])
        return winners.get(match_id, "")
    
    if label.startswith("Loser Match") and losers is not None:
        match_id = int(re.findall(r"\d+", label)[0])
        return losers.get(match_id, "")
    
    return canonical_team(label)

def play_match(team_a, team_b):
    p_a = get_advance_probability(team_a, team_b)
    return team_a if rng.random() < p_a else team_b

def deterministic_winner(team_a, team_b):
    p_a = get_advance_probability(team_a, team_b)
    
    if p_a >= 0.5:
        return team_a, p_a
    else:
        return team_b, 1 - p_a

round_sequence = ["r32", "r16", "qf", "sf", "final"]

round_probability_cols = {
    "r32": "round_32_probability",
    "r16": "round_16_probability",
    "qf": "quarterfinal_probability",
    "sf": "semifinal_probability",
    "final": "final_probability"
}

teams_in_model = sorted(ranking["team"].unique())

round_counts = {
    team: {
        "round_32_probability": 0,
        "round_16_probability": 0,
        "quarterfinal_probability": 0,
        "semifinal_probability": 0,
        "final_probability": 0,
        "champion_probability": 0
    }
    for team in teams_in_model
}

# =========================
# 7. Simulación Monte Carlo
# =========================

matches_to_simulate = knockout[
    knockout["round_type"].isin(round_sequence)
].copy()

matches_to_simulate = matches_to_simulate.sort_values("match_id")

for _ in range(N_SIMULATIONS):
    winners = {}
    losers = {}
    
    # Contar equipos que arrancan en 16avos
    r32_matches = matches_to_simulate[matches_to_simulate["round_type"] == "r32"]
    
    for _, row in r32_matches.iterrows():
        team_a = resolve_participant(row["team_a"], winners, losers)
        team_b = resolve_participant(row["team_b"], winners, losers)
        
        if team_a in round_counts:
            round_counts[team_a]["round_32_probability"] += 1
        
        if team_b in round_counts:
            round_counts[team_b]["round_32_probability"] += 1
    
    for _, row in matches_to_simulate.iterrows():
        match_id = int(row["match_id"])
        round_type = row["round_type"]
        
        team_a = resolve_participant(row["team_a"], winners, losers)
        team_b = resolve_participant(row["team_b"], winners, losers)
        
        # Si ya hay ganador real, se respeta
        real_winner = canonical_team(row.get("real_winner", ""))
        finished = bool(row.get("finished", False))
        
        if finished and real_winner and real_winner != "Draw":
            winner = real_winner
        else:
            winner = play_match(team_a, team_b)
        
        loser = team_b if winner == team_a else team_a
        
        winners[match_id] = winner
        losers[match_id] = loser
        
        if round_type == "r32":
            if winner in round_counts:
                round_counts[winner]["round_16_probability"] += 1
        
        elif round_type == "r16":
            if winner in round_counts:
                round_counts[winner]["quarterfinal_probability"] += 1
        
        elif round_type == "qf":
            if winner in round_counts:
                round_counts[winner]["semifinal_probability"] += 1
        
        elif round_type == "sf":
            if winner in round_counts:
                round_counts[winner]["final_probability"] += 1
        
        elif round_type == "final":
            if winner in round_counts:
                round_counts[winner]["champion_probability"] += 1

# =========================
# 8. Probabilidades por equipo
# =========================

round_probabilities = []

for team, counts in round_counts.items():
    row = {
        "team": team,
        "team_display": team_display(team),
        "confederation": team_confederation(team),
        "ranking": team_ranking(team),
        "power_score_live": team_power(team),
        "rating": team_rating(team)
    }
    
    for col, value in counts.items():
        row[col] = value / N_SIMULATIONS
    
    round_probabilities.append(row)

round_probabilities = pd.DataFrame(round_probabilities)

round_probabilities = round_probabilities.sort_values(
    "champion_probability",
    ascending=False
).reset_index(drop=True)

top_candidates = round_probabilities.head(8).copy()

confederation_probabilities = (
    round_probabilities
    .groupby("confederation", as_index=False)["champion_probability"]
    .sum()
    .sort_values("champion_probability", ascending=False)
)

surprises = (
    round_probabilities[
        round_probabilities["ranking"] > 8
    ]
    .sort_values("champion_probability", ascending=False)
    .head(5)
    .copy()
)

# =========================
# 9. Bracket proyectado determinístico real
# =========================

projected_rows = []
projected_winners = {}
projected_losers = {}

for _, row in matches_to_simulate.iterrows():
    match_id = int(row["match_id"])
    round_type = row["round_type"]
    
    team_a = resolve_participant(row["team_a"], projected_winners, projected_losers)
    team_b = resolve_participant(row["team_b"], projected_winners, projected_losers)
    
    real_winner = canonical_team(row.get("real_winner", ""))
    finished = bool(row.get("finished", False))
    
    if finished and real_winner and real_winner != "Draw":
        winner = real_winner
        winner_probability = 1.0
    else:
        winner, winner_probability = deterministic_winner(team_a, team_b)
    
    loser = team_b if winner == team_a else team_a
    
    projected_winners[match_id] = winner
    projected_losers[match_id] = loser
    
    projected_rows.append({
        "match_id": match_id,
        "round_type": round_type,
        "date": row.get("date", ""),
        "team_a": team_a,
        "team_b": team_b,
        "team_a_display": team_display(team_a),
        "team_b_display": team_display(team_b),
        "projected_winner": winner,
        "projected_winner_display": team_display(winner),
        "winner_probability": winner_probability,
        "next_match_id": row.get("next_match_id", ""),
        "bracket_side": row.get("bracket_side", "")
    })

projected_bracket = pd.DataFrame(projected_rows)

final_projected = projected_bracket[projected_bracket["round_type"] == "final"].iloc[0]
champion_projected = final_projected["projected_winner"]

champion_probs = round_probabilities[
    round_probabilities["team"] == champion_projected
].iloc[0]

round_progress_champion = pd.DataFrame([
    {
        "round": "16avos",
        "probability": champion_probs["round_32_probability"]
    },
    {
        "round": "Octavos",
        "probability": champion_probs["round_16_probability"]
    },
    {
        "round": "Cuartos",
        "probability": champion_probs["quarterfinal_probability"]
    },
    {
        "round": "Semifinales",
        "probability": champion_probs["semifinal_probability"]
    },
    {
        "round": "Final",
        "probability": champion_probs["final_probability"]
    },
    {
        "round": "Campeón",
        "probability": champion_probs["champion_probability"]
    }
])

sf_projected = projected_bracket[projected_bracket["round_type"] == "sf"].copy()

key_cards = pd.DataFrame([{
    "champion": champion_projected,
    "champion_display": team_display(champion_projected),
    "champion_probability": champion_probs["champion_probability"],
    "final_probability": champion_probs["final_probability"],
    "semifinal_probability": champion_probs["semifinal_probability"],
    "teams_simulated": len(teams_in_model),
    "simulations": N_SIMULATIONS,
    "final_team_a_display": final_projected["team_a_display"],
    "final_team_b_display": final_projected["team_b_display"],
    "final_winner_display": final_projected["projected_winner_display"],
    "final_match_probability": final_projected["winner_probability"],
    "semi_1_team_a_display": sf_projected.iloc[0]["team_a_display"],
    "semi_1_team_b_display": sf_projected.iloc[0]["team_b_display"],
    "semi_1_winner_display": sf_projected.iloc[0]["projected_winner_display"],
    "semi_1_probability": sf_projected.iloc[0]["winner_probability"],
    "semi_2_team_a_display": sf_projected.iloc[1]["team_a_display"],
    "semi_2_team_b_display": sf_projected.iloc[1]["team_b_display"],
    "semi_2_winner_display": sf_projected.iloc[1]["projected_winner_display"],
    "semi_2_probability": sf_projected.iloc[1]["winner_probability"],
}])

# =========================
# 10. Exportar para Power BI
# =========================

output_file = bracket_file.parent / "world_cup_tournament_simulation_real_bracket.xlsx"

with pd.ExcelWriter(output_file, engine="openpyxl") as writer:
    key_cards.to_excel(writer, sheet_name="key_cards_real", index=False)
    round_probabilities.to_excel(writer, sheet_name="round_probabilities_real", index=False)
    top_candidates.to_excel(writer, sheet_name="top_candidates_real", index=False)
    round_progress_champion.to_excel(writer, sheet_name="round_progress_champion_real", index=False)
    confederation_probabilities.to_excel(writer, sheet_name="confederation_probabilities_real", index=False)
    surprises.to_excel(writer, sheet_name="surprises_real", index=False)
    projected_bracket.to_excel(writer, sheet_name="projected_bracket_real", index=False)
    third_place_assignments.to_excel(writer, sheet_name="third_place_assignments", index=False)

print("Archivo creado correctamente:")
print(output_file)

print("\nCampeón proyectado:")
display(key_cards)

print("\nTop 8 candidatos:")
display(top_candidates[[
    "team_display",
    "champion_probability",
    "final_probability",
    "semifinal_probability",
    "confederation"
]])

print("\nBracket proyectado real:")
display(projected_bracket)
```

```python
import pandas as pd
import numpy as np
import re
import unicodedata
from pathlib import Path

print("Corrigiendo coherencia del campeón proyectado con favorito global...")

UNCERTAINTY_MARGIN = 0.02
# 0.02 = margen de 2 puntos porcentuales.
# Si un partido está entre 48% y 52%, gana el equipo con mayor probabilidad global de campeón.

# =========================
# 1. Buscar archivos
# =========================

search_roots = [
    Path.cwd(),
    Path.cwd().parent,
    Path.cwd().parent.parent
]

bracket_files = []
ranking_files = []
simulator_files = []
tournament_files = []

for root in search_roots:
    if root.exists():
        bracket_files.extend(list(root.rglob("official_tournament_bracket_normalized.xlsx")))
        ranking_files.extend(list(root.rglob("world_cup_live_power_ranking.xlsx")))
        simulator_files.extend(list(root.rglob("world_cup_match_simulator.xlsx")))
        tournament_files.extend(list(root.rglob("world_cup_tournament_simulation_real_bracket.xlsx")))

bracket_files = list(dict.fromkeys(bracket_files))
ranking_files = list(dict.fromkeys(ranking_files))
simulator_files = list(dict.fromkeys(simulator_files))
tournament_files = list(dict.fromkeys(tournament_files))

if not bracket_files:
    raise FileNotFoundError("No encontré official_tournament_bracket_normalized.xlsx")

if not ranking_files:
    raise FileNotFoundError("No encontré world_cup_live_power_ranking.xlsx")

if not simulator_files:
    raise FileNotFoundError("No encontré world_cup_match_simulator.xlsx")

if not tournament_files:
    raise FileNotFoundError("No encontré world_cup_tournament_simulation_real_bracket.xlsx")

bracket_file = bracket_files[0]
ranking_file = ranking_files[0]
simulator_file = simulator_files[0]
tournament_file = tournament_files[0]

print("Bracket real:", bracket_file)
print("Ranking:", ranking_file)
print("Simulador:", simulator_file)
print("Simulación real:", tournament_file)

# =========================
# 2. Cargar datos
# =========================

bracket_sheets = pd.read_excel(bracket_file, sheet_name=None)
tournament_sheets = pd.read_excel(tournament_file, sheet_name=None)

knockout = bracket_sheets["knockout_bracket"].copy()
ranking = pd.read_excel(ranking_file)
simulator = pd.read_excel(simulator_file)

round_probabilities = tournament_sheets["round_probabilities_real"].copy()

# =========================
# 3. Normalización de nombres
# =========================

name_aliases = {
    "Curaçao": "Curacao",
    "Curacao": "Curacao",
    "Czechia": "Czech Republic",
    "Czech Republic": "Czech Republic",
    "Democratic Republic of the Congo": "Congo",
    "DR Congo": "Congo",
    "Congo DR": "Congo",
    "Congo": "Congo",
    "United States": "United States",
    "USA": "United States",
}

def normalize_text(value):
    if pd.isna(value):
        return ""
    value = str(value).strip()
    value = unicodedata.normalize("NFKD", value)
    value = "".join([c for c in value if not unicodedata.combining(c)])
    return value

def canonical_team(value):
    if pd.isna(value):
        return ""
    
    value = str(value).strip()
    
    if value in name_aliases:
        return name_aliases[value]
    
    value_clean = normalize_text(value)
    
    for original, mapped in name_aliases.items():
        if normalize_text(original).lower() == value_clean.lower():
            return mapped
    
    return value

for col in ["team", "team_display"]:
    if col in round_probabilities.columns:
        round_probabilities[col] = round_probabilities[col].apply(
            lambda x: x if col == "team_display" else canonical_team(x)
        )

ranking["team"] = ranking["team"].apply(canonical_team)

if "team_display" not in ranking.columns:
    ranking["team_display"] = ranking["team"]

ranking = ranking.drop_duplicates(subset=["team"]).reset_index(drop=True)

team_info = ranking.set_index("team").to_dict("index")

def team_display(team):
    team = canonical_team(team)
    return team_info.get(team, {}).get("team_display", team)

def team_confederation(team):
    team = canonical_team(team)
    return team_info.get(team, {}).get("confederation", "")

def team_ranking(team):
    team = canonical_team(team)
    return team_info.get(team, {}).get("ranking", 99)

def team_power(team):
    team = canonical_team(team)
    return float(team_info.get(team, {}).get("power_score_live", 0.50))

def team_rating(team):
    team = canonical_team(team)
    return float(team_info.get(team, {}).get("rating", 1500))

# =========================
# 4. Favorito global del modelo
# =========================

round_probabilities["team"] = round_probabilities["team"].apply(canonical_team)

round_probabilities = round_probabilities.sort_values(
    "champion_probability",
    ascending=False
).reset_index(drop=True)

global_favorite_row = round_probabilities.iloc[0]
global_favorite = global_favorite_row["team"]

champion_prob_map = dict(
    zip(
        round_probabilities["team"],
        round_probabilities["champion_probability"]
    )
)

print("Favorito global del modelo:", team_display(global_favorite))
print("Probabilidad global:", round(global_favorite_row["champion_probability"] * 100, 2), "%")

# =========================
# 5. Aplicar terceros ya resueltos
# =========================

if "third_place_assignments" in tournament_sheets:
    third_place_assignments = tournament_sheets["third_place_assignments"].copy()
    
    if not third_place_assignments.empty:
        for _, row in third_place_assignments.iterrows():
            match_id = int(row["match_id"])
            side = str(row["side"])
            selected_team = canonical_team(row["selected_team"])
            
            team_col = f"team_{side}"
            status_col = f"team_{side}_status"
            
            mask = knockout["match_id"] == match_id
            
            if team_col in knockout.columns:
                knockout.loc[mask, team_col] = selected_team
            
            if status_col in knockout.columns:
                knockout.loc[mask, status_col] = "resolved_third_place_model"

# =========================
# 6. Probabilidades de cruce directo
# =========================

for col in ["team_a", "team_b"]:
    simulator[col] = simulator[col].apply(canonical_team)

for col in ["win_a_probability", "draw_probability", "win_b_probability"]:
    simulator[col] = pd.to_numeric(simulator[col], errors="coerce")
    if simulator[col].max() > 1:
        simulator[col] = simulator[col] / 100

pair_probs = (
    simulator[
        ["team_a", "team_b", "win_a_probability", "draw_probability", "win_b_probability"]
    ]
    .drop_duplicates(subset=["team_a", "team_b"])
    .copy()
)

advance_prob = {}

for _, row in pair_probs.iterrows():
    a = row["team_a"]
    b = row["team_b"]
    win_a = float(row["win_a_probability"])
    draw = float(row["draw_probability"])
    win_b = float(row["win_b_probability"])
    
    advance_prob[(a, b)] = win_a + draw * 0.5
    advance_prob[(b, a)] = win_b + draw * 0.5

def fallback_probability(team_a, team_b):
    power_diff = team_power(team_a) - team_power(team_b)
    rating_diff = (team_rating(team_a) - team_rating(team_b)) / 600
    
    x = power_diff * 2.5 + rating_diff
    p = 1 / (1 + np.exp(-x))
    
    return float(np.clip(p, 0.05, 0.95))

def get_advance_probability(team_a, team_b):
    team_a = canonical_team(team_a)
    team_b = canonical_team(team_b)
    
    if team_a == team_b:
        return 1.0
    
    if (team_a, team_b) in advance_prob:
        return advance_prob[(team_a, team_b)]
    
    return fallback_probability(team_a, team_b)

# =========================
# 7. Resolver participantes
# =========================

def resolve_participant(label, winners, losers=None):
    label = str(label)
    
    label_clean = label.lower()
    
    if "winner" in label_clean and "match" in label_clean:
        nums = re.findall(r"\d+", label)
        if nums:
            match_id = int(nums[-1])
            return winners.get(match_id, "")
    
    if "loser" in label_clean and "match" in label_clean and losers is not None:
        nums = re.findall(r"\d+", label)
        if nums:
            match_id = int(nums[-1])
            return losers.get(match_id, "")
    
    return canonical_team(label)

# =========================
# 8. Regla de ganador coherente
# =========================

def projected_winner_consistent(team_a, team_b):
    team_a = canonical_team(team_a)
    team_b = canonical_team(team_b)
    
    p_a = get_advance_probability(team_a, team_b)
    p_b = 1 - p_a
    
    prob_a_global = champion_prob_map.get(team_a, 0)
    prob_b_global = champion_prob_map.get(team_b, 0)
    
    # Si el cruce está muy parejo, usamos probabilidad global de campeón como desempate.
    if abs(p_a - 0.5) <= UNCERTAINTY_MARGIN:
        if prob_a_global >= prob_b_global:
            return team_a, p_a, "global_favorite_tiebreak"
        else:
            return team_b, p_b, "global_favorite_tiebreak"
    
    # Si no está parejo, gana el que tiene mayor probabilidad directa del cruce.
    if p_a >= p_b:
        return team_a, p_a, "head_to_head_probability"
    else:
        return team_b, p_b, "head_to_head_probability"

# =========================
# 9. Recalcular bracket proyectado
# =========================

knockout["match_id"] = pd.to_numeric(knockout["match_id"], errors="coerce").astype(int)

round_sequence = ["r32", "r16", "qf", "sf", "final"]

matches_to_project = knockout[
    knockout["round_type"].isin(round_sequence)
].copy()

matches_to_project = matches_to_project.sort_values("match_id")

projected_rows = []
projected_winners = {}
projected_losers = {}

for _, row in matches_to_project.iterrows():
    match_id = int(row["match_id"])
    round_type = row["round_type"]
    
    team_a = resolve_participant(row["team_a"], projected_winners, projected_losers)
    team_b = resolve_participant(row["team_b"], projected_winners, projected_losers)
    
    real_winner = canonical_team(row.get("real_winner", ""))
    finished = bool(row.get("finished", False))
    
    if finished and real_winner and real_winner != "Draw":
        winner = real_winner
        winner_probability = 1.0
        decision_rule = "real_result"
    else:
        winner, winner_probability, decision_rule = projected_winner_consistent(team_a, team_b)
    
    loser = team_b if winner == team_a else team_a
    
    projected_winners[match_id] = winner
    projected_losers[match_id] = loser
    
    projected_rows.append({
        "match_id": match_id,
        "round_type": round_type,
        "date": row.get("date", ""),
        "team_a": team_a,
        "team_b": team_b,
        "team_a_display": team_display(team_a),
        "team_b_display": team_display(team_b),
        "projected_winner": winner,
        "projected_winner_display": team_display(winner),
        "winner_probability": winner_probability,
        "decision_rule": decision_rule,
        "next_match_id": row.get("next_match_id", ""),
        "bracket_side": row.get("bracket_side", "")
    })

projected_bracket = pd.DataFrame(projected_rows)

# =========================
# 10. Actualizar key cards
# =========================

final_projected = projected_bracket[projected_bracket["round_type"] == "final"].iloc[0]
champion_projected = final_projected["projected_winner"]

champion_probs = round_probabilities[
    round_probabilities["team"] == champion_projected
].iloc[0]

sf_projected = projected_bracket[projected_bracket["round_type"] == "sf"].copy()

round_progress_champion = pd.DataFrame([
    {
        "round": "16avos",
        "probability": champion_probs["round_32_probability"]
    },
    {
        "round": "Octavos",
        "probability": champion_probs["round_16_probability"]
    },
    {
        "round": "Cuartos",
        "probability": champion_probs["quarterfinal_probability"]
    },
    {
        "round": "Semifinales",
        "probability": champion_probs["semifinal_probability"]
    },
    {
        "round": "Final",
        "probability": champion_probs["final_probability"]
    },
    {
        "round": "Campeón",
        "probability": champion_probs["champion_probability"]
    }
])

previous_key_cards = tournament_sheets["key_cards_real"]

key_cards = pd.DataFrame([{
    "champion": champion_projected,
    "champion_display": team_display(champion_projected),
    "champion_probability": champion_probs["champion_probability"],
    "final_probability": champion_probs["final_probability"],
    "semifinal_probability": champion_probs["semifinal_probability"],
    "teams_simulated": previous_key_cards["teams_simulated"].iloc[0],
    "simulations": previous_key_cards["simulations"].iloc[0],
    "final_team_a_display": final_projected["team_a_display"],
    "final_team_b_display": final_projected["team_b_display"],
    "final_winner_display": final_projected["projected_winner_display"],
    "final_match_probability": final_projected["winner_probability"],
    "final_decision_rule": final_projected["decision_rule"],
    "semi_1_team_a_display": sf_projected.iloc[0]["team_a_display"],
    "semi_1_team_b_display": sf_projected.iloc[0]["team_b_display"],
    "semi_1_winner_display": sf_projected.iloc[0]["projected_winner_display"],
    "semi_1_probability": sf_projected.iloc[0]["winner_probability"],
    "semi_1_decision_rule": sf_projected.iloc[0]["decision_rule"],
    "semi_2_team_a_display": sf_projected.iloc[1]["team_a_display"],
    "semi_2_team_b_display": sf_projected.iloc[1]["team_b_display"],
    "semi_2_winner_display": sf_projected.iloc[1]["projected_winner_display"],
    "semi_2_probability": sf_projected.iloc[1]["winner_probability"],
    "semi_2_decision_rule": sf_projected.iloc[1]["decision_rule"],
}])

model_decision_notes = pd.DataFrame([
    {
        "rule": "global_favorite_tiebreak",
        "description": "Si un cruce está dentro de un margen de 48%-52%, el modelo desempata usando la probabilidad global de campeón."
    },
    {
        "rule": "head_to_head_probability",
        "description": "Si el cruce no está parejo, gana el equipo con mayor probabilidad directa en ese partido."
    },
    {
        "rule": "real_result",
        "description": "Si el partido ya terminó, se respeta el resultado real."
    }
])

# =========================
# 11. Guardar archivo actualizado
# =========================

tournament_sheets["key_cards_real"] = key_cards
tournament_sheets["projected_bracket_real"] = projected_bracket
tournament_sheets["round_progress_champion_real"] = round_progress_champion
tournament_sheets["model_decision_notes"] = model_decision_notes

with pd.ExcelWriter(tournament_file, engine="openpyxl") as writer:
    for sheet_name, df in tournament_sheets.items():
        df.to_excel(writer, sheet_name=sheet_name, index=False)

print("Archivo actualizado correctamente:")
print(tournament_file)

print("\nCampeón proyectado coherente:")
display(key_cards)

print("\nFinal proyectada:")
display(
    projected_bracket[
        projected_bracket["round_type"] == "final"
    ][
        [
            "team_a_display",
            "team_b_display",
            "projected_winner_display",
            "winner_probability",
            "decision_rule"
        ]
    ]
)

print("\nSemifinales proyectadas:")
display(
    projected_bracket[
        projected_bracket["round_type"] == "sf"
    ][
        [
            "team_a_display",
            "team_b_display",
            "projected_winner_display",
            "winner_probability",
            "decision_rule"
        ]
    ]
)
```

```python
import pandas as pd
import numpy as np
import re
import unicodedata
from pathlib import Path

print("Corrigiendo bracket real y regenerando simulación completa...")

N_SIMULATIONS = 25000
RANDOM_SEED = 42
UNCERTAINTY_MARGIN = 0.02

rng = np.random.default_rng(RANDOM_SEED)

# =========================
# 1. Buscar archivos
# =========================

search_roots = [
    Path.cwd(),
    Path.cwd().parent,
    Path.cwd().parent.parent
]

bracket_files = []
ranking_files = []
simulator_files = []
tournament_files = []

for root in search_roots:
    if root.exists():
        bracket_files.extend(list(root.rglob("official_tournament_bracket_normalized.xlsx")))
        ranking_files.extend(list(root.rglob("world_cup_live_power_ranking.xlsx")))
        simulator_files.extend(list(root.rglob("world_cup_match_simulator.xlsx")))
        tournament_files.extend(list(root.rglob("world_cup_tournament_simulation_real_bracket.xlsx")))

bracket_files = list(dict.fromkeys(bracket_files))
ranking_files = list(dict.fromkeys(ranking_files))
simulator_files = list(dict.fromkeys(simulator_files))
tournament_files = list(dict.fromkeys(tournament_files))

if not bracket_files:
    raise FileNotFoundError("No encontré official_tournament_bracket_normalized.xlsx")

if not ranking_files:
    raise FileNotFoundError("No encontré world_cup_live_power_ranking.xlsx")

if not simulator_files:
    raise FileNotFoundError("No encontré world_cup_match_simulator.xlsx")

if not tournament_files:
    raise FileNotFoundError("No encontré world_cup_tournament_simulation_real_bracket.xlsx")

bracket_file = bracket_files[0]
ranking_file = ranking_files[0]
simulator_file = simulator_files[0]
tournament_file = tournament_files[0]

print("Bracket real:", bracket_file)
print("Ranking:", ranking_file)
print("Simulador:", simulator_file)
print("Simulación:", tournament_file)

# =========================
# 2. Cargar datos
# =========================

bracket_sheets = pd.read_excel(bracket_file, sheet_name=None)
tournament_sheets = pd.read_excel(tournament_file, sheet_name=None)

knockout = bracket_sheets["knockout_bracket"].copy()
official_games = bracket_sheets["official_games"].copy()

ranking = pd.read_excel(ranking_file)
simulator = pd.read_excel(simulator_file)

# =========================
# 3. Normalizar nombres
# =========================

name_aliases = {
    "Curaçao": "Curacao",
    "Curacao": "Curacao",
    "Czechia": "Czech Republic",
    "Czech Republic": "Czech Republic",
    "Democratic Republic of the Congo": "Congo",
    "DR Congo": "Congo",
    "Congo DR": "Congo",
    "Congo": "Congo",
    "United States": "United States",
    "USA": "United States",
}

def normalize_text(value):
    if pd.isna(value):
        return ""
    value = str(value).strip()
    value = unicodedata.normalize("NFKD", value)
    value = "".join([c for c in value if not unicodedata.combining(c)])
    return value

def canonical_team(value):
    if pd.isna(value):
        return ""
    
    value = str(value).strip()
    
    if value in name_aliases:
        return name_aliases[value]
    
    value_clean = normalize_text(value)
    
    for original, mapped in name_aliases.items():
        if normalize_text(original).lower() == value_clean.lower():
            return mapped
    
    return value

ranking["team"] = ranking["team"].apply(canonical_team)

if "team_display" not in ranking.columns:
    ranking["team_display"] = ranking["team"]

ranking = ranking.drop_duplicates(subset=["team"]).reset_index(drop=True)

team_info = ranking.set_index("team").to_dict("index")

def team_display(team):
    team = canonical_team(team)
    return team_info.get(team, {}).get("team_display", team)

def team_confederation(team):
    team = canonical_team(team)
    return team_info.get(team, {}).get("confederation", "")

def team_ranking(team):
    team = canonical_team(team)
    return team_info.get(team, {}).get("ranking", 99)

def team_power(team):
    team = canonical_team(team)
    return float(team_info.get(team, {}).get("power_score_live", 0.50))

def team_rating(team):
    team = canonical_team(team)
    return float(team_info.get(team, {}).get("rating", 1500))

# =========================
# 4. Corregir 16avos reales
# =========================

r32_real_matches = {
    73: ("South Africa", "Canada"),
    74: ("Germany", "Paraguay"),
    75: ("Netherlands", "Morocco"),
    76: ("Brazil", "Japan"),
    77: ("France", "Sweden"),
    78: ("Ivory Coast", "Norway"),
    79: ("Mexico", "Ecuador"),
    80: ("England", "Congo"),
    81: ("United States", "Bosnia and Herzegovina"),
    82: ("Belgium", "Senegal"),
    83: ("Portugal", "Croatia"),
    84: ("Spain", "Austria"),
    85: ("Switzerland", "Algeria"),
    86: ("Argentina", "Cape Verde"),
    87: ("Colombia", "Ghana"),
    88: ("Australia", "Egypt"),
}

for match_id, (team_a, team_b) in r32_real_matches.items():
    mask_ko = knockout["match_id"] == match_id
    mask_all = official_games["match_id"] == match_id
    
    knockout.loc[mask_ko, "team_a"] = canonical_team(team_a)
    knockout.loc[mask_ko, "team_b"] = canonical_team(team_b)
    knockout.loc[mask_ko, "team_a_status"] = "official_team"
    knockout.loc[mask_ko, "team_b_status"] = "official_team"
    
    official_games.loc[mask_all, "team_a"] = canonical_team(team_a)
    official_games.loc[mask_all, "team_b"] = canonical_team(team_b)
    official_games.loc[mask_all, "team_a_status"] = "official_team"
    official_games.loc[mask_all, "team_b_status"] = "official_team"

print("16avos corregidos:")
display(
    knockout[
        knockout["round_type"] == "r32"
    ][
        ["match_id", "round_type", "team_a", "team_b", "next_match_id", "bracket_side"]
    ]
)

# =========================
# 5. Probabilidades de cruce directo
# =========================

for col in ["team_a", "team_b"]:
    simulator[col] = simulator[col].apply(canonical_team)

for col in ["win_a_probability", "draw_probability", "win_b_probability"]:
    simulator[col] = pd.to_numeric(simulator[col], errors="coerce")
    if simulator[col].max() > 1:
        simulator[col] = simulator[col] / 100

pair_probs = (
    simulator[
        ["team_a", "team_b", "win_a_probability", "draw_probability", "win_b_probability"]
    ]
    .drop_duplicates(subset=["team_a", "team_b"])
    .copy()
)

advance_prob = {}

for _, row in pair_probs.iterrows():
    a = canonical_team(row["team_a"])
    b = canonical_team(row["team_b"])
    win_a = float(row["win_a_probability"])
    draw = float(row["draw_probability"])
    win_b = float(row["win_b_probability"])
    
    advance_prob[(a, b)] = win_a + draw * 0.5
    advance_prob[(b, a)] = win_b + draw * 0.5

def fallback_probability(team_a, team_b):
    power_diff = team_power(team_a) - team_power(team_b)
    rating_diff = (team_rating(team_a) - team_rating(team_b)) / 600
    
    x = power_diff * 2.5 + rating_diff
    p = 1 / (1 + np.exp(-x))
    
    return float(np.clip(p, 0.05, 0.95))

def get_advance_probability(team_a, team_b):
    team_a = canonical_team(team_a)
    team_b = canonical_team(team_b)
    
    if team_a == team_b:
        return 1.0
    
    if (team_a, team_b) in advance_prob:
        return advance_prob[(team_a, team_b)]
    
    return fallback_probability(team_a, team_b)

# =========================
# 6. Resolver participantes
# =========================

def resolve_participant(label, winners, losers=None):
    label = str(label)
    label_clean = label.lower()
    
    if "winner" in label_clean and "match" in label_clean:
        nums = re.findall(r"\d+", label)
        if nums:
            match_id = int(nums[-1])
            return winners.get(match_id, "")
    
    if "loser" in label_clean and "match" in label_clean and losers is not None:
        nums = re.findall(r"\d+", label)
        if nums:
            match_id = int(nums[-1])
            return losers.get(match_id, "")
    
    return canonical_team(label)

def play_match(team_a, team_b):
    p_a = get_advance_probability(team_a, team_b)
    return team_a if rng.random() < p_a else team_b

# =========================
# 7. Simulación Monte Carlo sobre bracket corregido
# =========================

round_sequence = ["r32", "r16", "qf", "sf", "final"]

matches_to_simulate = knockout[
    knockout["round_type"].isin(round_sequence)
].copy()

matches_to_simulate["match_id"] = pd.to_numeric(
    matches_to_simulate["match_id"],
    errors="coerce"
).astype(int)

matches_to_simulate = matches_to_simulate.sort_values("match_id")

teams_in_model = sorted(ranking["team"].unique())

round_counts = {
    team: {
        "round_32_probability": 0,
        "round_16_probability": 0,
        "quarterfinal_probability": 0,
        "semifinal_probability": 0,
        "final_probability": 0,
        "champion_probability": 0
    }
    for team in teams_in_model
}

print("Iniciando simulaciones...")

for sim in range(N_SIMULATIONS):
    if sim % 5000 == 0:
        print("Simulación:", sim)
    
    winners = {}
    losers = {}
    
    # Contar participantes reales de 16avos
    r32_matches = matches_to_simulate[
        matches_to_simulate["round_type"] == "r32"
    ]
    
    for _, row in r32_matches.iterrows():
        team_a = canonical_team(row["team_a"])
        team_b = canonical_team(row["team_b"])
        
        if team_a in round_counts:
            round_counts[team_a]["round_32_probability"] += 1
        
        if team_b in round_counts:
            round_counts[team_b]["round_32_probability"] += 1
    
    for _, row in matches_to_simulate.iterrows():
        match_id = int(row["match_id"])
        round_type = row["round_type"]
        
        team_a = resolve_participant(row["team_a"], winners, losers)
        team_b = resolve_participant(row["team_b"], winners, losers)
        
        real_winner = canonical_team(row.get("real_winner", ""))
        finished = bool(row.get("finished", False))
        
        if finished and real_winner and real_winner != "Draw":
            winner = real_winner
        else:
            winner = play_match(team_a, team_b)
        
        loser = team_b if winner == team_a else team_a
        
        winners[match_id] = winner
        losers[match_id] = loser
        
        if round_type == "r32":
            if winner in round_counts:
                round_counts[winner]["round_16_probability"] += 1
        
        elif round_type == "r16":
            if winner in round_counts:
                round_counts[winner]["quarterfinal_probability"] += 1
        
        elif round_type == "qf":
            if winner in round_counts:
                round_counts[winner]["semifinal_probability"] += 1
        
        elif round_type == "sf":
            if winner in round_counts:
                round_counts[winner]["final_probability"] += 1
        
        elif round_type == "final":
            if winner in round_counts:
                round_counts[winner]["champion_probability"] += 1

print("Simulaciones terminadas.")

# =========================
# 8. Tablas de probabilidades
# =========================

round_probabilities = []

for team, counts in round_counts.items():
    row = {
        "team": team,
        "team_display": team_display(team),
        "confederation": team_confederation(team),
        "ranking": team_ranking(team),
        "power_score_live": team_power(team),
        "rating": team_rating(team)
    }
    
    for col, value in counts.items():
        row[col] = value / N_SIMULATIONS
    
    round_probabilities.append(row)

round_probabilities = pd.DataFrame(round_probabilities)

round_probabilities = round_probabilities.sort_values(
    "champion_probability",
    ascending=False
).reset_index(drop=True)

top_candidates = round_probabilities.head(8).copy()

confederation_probabilities = (
    round_probabilities
    .groupby("confederation", as_index=False)["champion_probability"]
    .sum()
    .sort_values("champion_probability", ascending=False)
)

surprises = (
    round_probabilities[
        round_probabilities["ranking"] > 8
    ]
    .sort_values("champion_probability", ascending=False)
    .head(5)
    .copy()
)

champion_prob_map = dict(
    zip(
        round_probabilities["team"],
        round_probabilities["champion_probability"]
    )
)

# =========================
# 9. Bracket proyectado coherente
# =========================

def projected_winner_consistent(team_a, team_b):
    team_a = canonical_team(team_a)
    team_b = canonical_team(team_b)
    
    p_a = get_advance_probability(team_a, team_b)
    p_b = 1 - p_a
    
    prob_a_global = champion_prob_map.get(team_a, 0)
    prob_b_global = champion_prob_map.get(team_b, 0)
    
    # Si está dentro de 48%-52%, desempata la probabilidad global de campeón
    if abs(p_a - 0.5) <= UNCERTAINTY_MARGIN:
        if prob_a_global >= prob_b_global:
            return team_a, p_a, "global_favorite_tiebreak"
        else:
            return team_b, p_b, "global_favorite_tiebreak"
    
    if p_a >= p_b:
        return team_a, p_a, "head_to_head_probability"
    else:
        return team_b, p_b, "head_to_head_probability"

projected_rows = []
projected_winners = {}
projected_losers = {}

for _, row in matches_to_simulate.iterrows():
    match_id = int(row["match_id"])
    round_type = row["round_type"]
    
    team_a = resolve_participant(row["team_a"], projected_winners, projected_losers)
    team_b = resolve_participant(row["team_b"], projected_winners, projected_losers)
    
    real_winner = canonical_team(row.get("real_winner", ""))
    finished = bool(row.get("finished", False))
    
    if finished and real_winner and real_winner != "Draw":
        winner = real_winner
        winner_probability = 1.0
        decision_rule = "real_result"
    else:
        winner, winner_probability, decision_rule = projected_winner_consistent(team_a, team_b)
    
    loser = team_b if winner == team_a else team_a
    
    projected_winners[match_id] = winner
    projected_losers[match_id] = loser
    
    projected_rows.append({
        "match_id": match_id,
        "round_type": round_type,
        "date": row.get("date", ""),
        "team_a": team_a,
        "team_b": team_b,
        "team_a_display": team_display(team_a),
        "team_b_display": team_display(team_b),
        "projected_winner": winner,
        "projected_winner_display": team_display(winner),
        "winner_probability": winner_probability,
        "decision_rule": decision_rule,
        "next_match_id": row.get("next_match_id", ""),
        "bracket_side": row.get("bracket_side", "")
    })

projected_bracket = pd.DataFrame(projected_rows)

# =========================
# 10. Key cards
# =========================

final_projected = projected_bracket[
    projected_bracket["round_type"] == "final"
].iloc[0]

champion_projected = final_projected["projected_winner"]

champion_probs = round_probabilities[
    round_probabilities["team"] == champion_projected
].iloc[0]

sf_projected = projected_bracket[
    projected_bracket["round_type"] == "sf"
].copy()

round_progress_champion = pd.DataFrame([
    {"round": "16avos", "probability": champion_probs["round_32_probability"]},
    {"round": "Octavos", "probability": champion_probs["round_16_probability"]},
    {"round": "Cuartos", "probability": champion_probs["quarterfinal_probability"]},
    {"round": "Semifinales", "probability": champion_probs["semifinal_probability"]},
    {"round": "Final", "probability": champion_probs["final_probability"]},
    {"round": "Campeón", "probability": champion_probs["champion_probability"]},
])

key_cards = pd.DataFrame([{
    "champion": champion_projected,
    "champion_display": team_display(champion_projected),
    "champion_probability": champion_probs["champion_probability"],
    "final_probability": champion_probs["final_probability"],
    "semifinal_probability": champion_probs["semifinal_probability"],
    "teams_simulated": len(teams_in_model),
    "simulations": N_SIMULATIONS,
    "final_team_a_display": final_projected["team_a_display"],
    "final_team_b_display": final_projected["team_b_display"],
    "final_winner_display": final_projected["projected_winner_display"],
    "final_match_probability": final_projected["winner_probability"],
    "final_decision_rule": final_projected["decision_rule"],
    "semi_1_team_a_display": sf_projected.iloc[0]["team_a_display"],
    "semi_1_team_b_display": sf_projected.iloc[0]["team_b_display"],
    "semi_1_winner_display": sf_projected.iloc[0]["projected_winner_display"],
    "semi_1_probability": sf_projected.iloc[0]["winner_probability"],
    "semi_1_decision_rule": sf_projected.iloc[0]["decision_rule"],
    "semi_2_team_a_display": sf_projected.iloc[1]["team_a_display"],
    "semi_2_team_b_display": sf_projected.iloc[1]["team_b_display"],
    "semi_2_winner_display": sf_projected.iloc[1]["projected_winner_display"],
    "semi_2_probability": sf_projected.iloc[1]["winner_probability"],
    "semi_2_decision_rule": sf_projected.iloc[1]["decision_rule"],
}])

model_decision_notes = pd.DataFrame([
    {
        "rule": "global_favorite_tiebreak",
        "description": "Si un cruce está dentro de 48%-52%, el modelo desempata usando la probabilidad global de campeón."
    },
    {
        "rule": "head_to_head_probability",
        "description": "Si el cruce no está parejo, gana el equipo con mayor probabilidad directa."
    },
    {
        "rule": "real_result",
        "description": "Si el partido ya terminó, se respeta el resultado real."
    }
])

# =========================
# 11. Guardar archivos
# =========================

# Actualizar archivo de bracket normalizado
bracket_sheets["official_games"] = official_games
bracket_sheets["knockout_bracket"] = knockout

with pd.ExcelWriter(bracket_file, engine="openpyxl") as writer:
    for sheet_name, df in bracket_sheets.items():
        df.to_excel(writer, sheet_name=sheet_name, index=False)

# Actualizar archivo de simulación
tournament_sheets["key_cards_real"] = key_cards
tournament_sheets["round_probabilities_real"] = round_probabilities
tournament_sheets["top_candidates_real"] = top_candidates
tournament_sheets["round_progress_champion_real"] = round_progress_champion
tournament_sheets["confederation_probabilities_real"] = confederation_probabilities
tournament_sheets["surprises_real"] = surprises
tournament_sheets["projected_bracket_real"] = projected_bracket
tournament_sheets["model_decision_notes"] = model_decision_notes

with pd.ExcelWriter(tournament_file, engine="openpyxl") as writer:
    for sheet_name, df in tournament_sheets.items():
        df.to_excel(writer, sheet_name=sheet_name, index=False)

print("Archivos actualizados correctamente.")
print("Bracket:", bracket_file)
print("Simulación:", tournament_file)

print("\nCampeón proyectado:")
display(key_cards)

print("\nTop candidatos:")
display(
    top_candidates[
        ["team_display", "champion_probability", "final_probability", "semifinal_probability"]
    ]
)

print("\nBracket proyectado corregido:")
display(
    projected_bracket[
        ["match_id", "round_type", "team_a_display", "team_b_display", "projected_winner_display", "winner_probability", "decision_rule", "next_match_id", "bracket_side"]
    ]
)
```
